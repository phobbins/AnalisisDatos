## Pregunta 6

```python
import numpy as np

reclamaciones = [13, 15, 8, 16, 8]

#A) Calcule el número medio de reclamaciones semanales
print("Media:",np.mean(reclamaciones))

#B) Calcule el número mediano de reclamaciones semanales
print("Mediana:", np.median(reclamaciones))

#C) Halle la moda
valores, conteos = np.unique(reclamaciones, return_counts=True)

# Encontrar el valor con la mayor frecuencia
moda = valores[np.argmax(conteos)]

print("Moda:", moda)
```

## Pregunta 11

Considere la siguiente distribución de frecuencias de una muestra de 40
observaciones:


Clase  Frecuencia


 0-4     5
 
 5-9     8
 
 10-14   11
 
 15-19   9
 
 20-24   7
 
a) Calcule la media muestral.

b) Calcule la varianza muestral y la desviación estándar muestral.

```python
import numpy as np

fi = np.array([5,8,11,9,7])

lim_inf = np.array([0,5,10,15,20])
lim_sup = np.array([4,9,14,19,24])


#A)
puntos_medios = (lim_inf + lim_sup) / 2

media_muestral = np.sum(fi * puntos_medios) / np.sum(fi)

print('La media muestral es',media_muestral)

#B) Varianza muestral
var = np.sum(fi*(puntos_medios - media_muestral)**2) / np.sum(fi)
print('La varianza muestral ',var)

#B) Desviacion muestral
print('La desviacion estandar ', np.sqrt(var))
```

## Pregunta 16

16.Un hospital tiene interés en averiguar la eficacia de un nuevo medicamento
para reducir el tiempo necesario para recuperarse totalmente de una
operación de rodilla. La recuperación total se mide por medio de una serie de
tests de fuerza que comparan la rodilla operada con la rodilla sin operar. El
medicamento se administró en dosis diferentes a 18 pacientes durante un
periodo de 6 meses. Los datos (x, y) siguientes indican el número de unidades
de medicamento, X, y los días necesarios para la recuperación total Y de cada
paciente:

(5, 53) (4, 56) (9, 54) (14, 56) (21, 65) (7, 53) (7, 56) (9, 59) (14, 48) (11, 66)
(21,57) (17, 49) (9, 53) (21, 52) (4, 56) (9, 46) (14, 66) (13, 49)

a) Calcule la covarianza.

b) Calcule el coeficiente de correlación.

c) Analice brevemente la relación entre el número de unidades de
medicamento y el tiempo de recuperación. ¿Qué dosis deberíamos
recomendar basándonos en este análisis inicial?

```python
import numpy as np

x = np.array([5, 4, 9, 14, 21, 7, 7, 9, 14, 11, 21, 17, 9, 21, 4, 9, 14, 13])
y = np.array([53, 56, 54, 56, 65, 53, 56, 59, 48, 66, 57, 49, 53, 52, 56, 46, 66, 49])

#A) Covarianza
varXdeY = np.cov(x,y)
print("Covarianza: ",varXdeY[0,1])

#B)Calcule el coeficiente de correlación.
std_x = np.std(x, ddof=1)
std_y = np.std(y, ddof=1)

corr = varXdeY[0,1]/(std_x*std_y)
print("Correlacion: ", corr)
```

Vemos que la covarianza es 4.26, por lo que hay una relacion entre los medicamentos y los dias de recuperacion. Pero por otro lado la correlacion es de 0.128, asi que la relacion es muy pequeña. Si vemos el grafico siguiente notamos como los datos estan dispersos.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Datos', marker='o')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label='Línea de Regresión')

# Etiquetas y título
plt.title('Dosis de Medicamento vs. Tiempo de Recuperación')
plt.xlabel('Dosis de Medicamento (x)')
plt.ylabel('Tiempo de Recuperación (y)')
plt.legend()
plt.grid()

plt.show()
```

## Pregunta 21

Un concesionario de automóviles calcula la proporción de automóviles
nuevos vendidos que se han devuelto varias veces para que se corrijan los
defectos durante el periodo de garantía. Los resultados fueron los siguientes:

Número de devoluciones 0 1 2 3 4

Proporción 0,28 0,36 0,23 0,09 0,04

a) Graficar la función de probabilidad.

b) Calcular y graficar la función de probabilidad acumulada.

c) Halle la esperanza del número de devoluciones de un automóvil para que se
corrijan los defectos durante el periodo de garantía.

d) Halle la varianza del número de devoluciones de un automóvil para que se
corrijan los defectos durante el periodo de garantía.


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
devoluciones = np.array([0, 1, 2, 3, 4])
proporciones = np.array([0.28, 0.36, 0.23, 0.09, 0.04])

# A) Función de probabilidad
plt.figure(figsize=(10, 6))
plt.bar(devoluciones, proporciones, color='skyblue', width=0.5)

# Etiquetas y título
plt.title('Función de Probabilidad de Devoluciones de Automóviles Nuevos')
plt.xlabel('Número de Devoluciones')
plt.ylabel('Proporción')
plt.xticks(devoluciones)  
plt.ylim(0, 0.4)  


plt.grid(axis='y')
plt.show()

# B) Probabilidad acumulada
probabilidad_acumulada = np.cumsum(proporciones)

plt.figure(figsize=(10, 6))
plt.bar(devoluciones, probabilidad_acumulada, color='skyblue', width=0.5)


plt.title('Función de Probabilidad Acumulada de Devoluciones de Automóviles Nuevos')
plt.xlabel('Número de Devoluciones')
plt.ylabel('Probabilidad Acumulada')
plt.xticks(devoluciones)  
plt.ylim(0, 1) 

# Mostrar el gráfico
plt.grid(axis='y')
plt.show()




#C) Cálculo de la esperanza
esperanza = np.sum(devoluciones * proporciones)
print('La esperanza es', esperanza)

#D) Varianza, Var E(X^2) - [E(X)]^2
esperanza = np.sum(devoluciones * proporciones)
# Cálculo de E(X^2)
esperanza_cuadrado = np.sum((devoluciones ** 2) * proporciones)
varianza = esperanza_cuadrado - (esperanza ** 2)
print("La varianza del número de devoluciones es:", varianza)
```
