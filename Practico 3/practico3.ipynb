### Ejercicio 10

```python
import numpy as np

# Datos de la muestra (Xi) y ponderaciones (Wi)
xi = [4.6, 3.2, 5.4, 2.6, 5.2]
wi = [8, 3, 6, 2, 5]

# a) Cálculo de la media aritmética sin ponderaciones
media_aritmetica = np.mean(xi)

# b) Cálculo de la media ponderada
media_ponderada = np.average(xi, weights=wi)

# Resultados
print(f"Media aritmética sin ponderaciones: {media_aritmetica}")
print(f"Media ponderada: {media_ponderada}")
```

### Ejercicio 15

```python
import numpy as np

# Datos de la muestra: pares (x, y)
data = [(1, 5), (3, 7), (4, 6), (3, 6), (5, 7), (5, 8)]

# Separar los pares en listas de x y y
x, y = zip(*data)

# a) Cálculo de la covarianza
covarianza = np.cov(x, y)[0][1]

# b) Cálculo del coeficiente de correlación
coeficiente_correlacion = np.corrcoef(x, y)[0][1]

# Resultados
print(f"Covarianza: {covarianza}")
print(f"Coeficiente de correlación: {coeficiente_correlacion}")
```

### Ejercicio 20

```python
# Distribución de probabilidad
X = [0, 1, 2, 3, 4, 5]
P = [0.10, 0.15, 0.20, 0.25, 0.20, 0.10]

# a) Probabilidad de que lleguen al menos 2 personas
P_al_menos_2 = sum(P[2:])  # Suma de P(2), P(3), P(4), P(5)

# b) Probabilidad de que llegue más de 3 personas
P_mas_de_3 = sum(P[4:])  # Suma de P(4), P(5)

# c) Probabilidad de que llegue como máximo 4 personas
P_como_maximo_4 = sum(P[:5])  # Suma de P(0), P(1), P(2), P(3), P(4)

# d) Probabilidad de que lleguen entre 1 y 3 personas
P_entre_1_y_3 = sum(P[1:4])  # Suma de P(1), P(2), P(3)

# Resultados
print(f"Probabilidad de que lleguen al menos 2 personas: {P_al_menos_2}")
print(f"Probabilidad de que llegue más de 3 personas: {P_mas_de_3}")
print(f"Probabilidad de que llegue como máximo 4 personas: {P_como_maximo_4}")
print(f"Probabilidad de que lleguen entre 1 y 3 personas: {P_entre_1_y_3}")
```

### Ejercicio 22

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la función de probabilidad
pedidos = np.array([0, 1, 2, 3, 4, 5])
probabilidades = np.array([0.10, 0.14, 0.26, 0.28, 0.15, 0.07])

# a) Graficar la función de probabilidad
plt.figure(figsize=(8, 5))
plt.bar(pedidos, probabilidades, color='skyblue')
# Añadir el valor de la probabilidad dentro de las barras
for i in range(len(pedidos)):
    plt.text(pedidos[i], probabilidades[i], f"{probabilidades[i]:.2f}", ha='center', va='bottom')
plt.xlabel('Número de pedidos')
plt.ylabel('Probabilidad')
plt.title('Función de probabilidad')
plt.show()

# b) Calcular y graficar la función de probabilidad acumulada
prob_acumulada = np.cumsum(probabilidades)  # Calcular la probabilidad acumulada
plt.figure(figsize=(8, 5))
plt.step(pedidos, prob_acumulada, where='post', color='skyblue', linewidth=2, label="Probabilidad acumulada")
plt.scatter(pedidos, prob_acumulada, color='skyblue')  # Añadir puntos para mayor claridad
plt.xlabel('Número de pedidos')
plt.ylabel('Probabilidad acumulada')
plt.title('Función de probabilidad acumulada')
plt.grid(True, linestyle='--')
plt.legend()
plt.xlim(0, 5.5)
plt.ylim(0, 1.1)
plt.show()

# c) Probabilidad de que se hagan al menos 3 pedidos (P(X >= 3))
prob_al_menos_3 = np.sum(probabilidades[pedidos >= 3])
print(f"La probabilidad de que se hagan al menos 3 pedidos es de {prob_al_menos_3}")

# d) Calcular la media
media = np.sum(pedidos * probabilidades)
print(f"La media es de {media}")

# e) Calcular la desviación típica
varianza = np.sum(((pedidos - media) ** 2) * probabilidades)
desviacion_tipica = np.sqrt(varianza)
print(f"La desviación típica es de {desviacion_tipica}")
```

### Ejercicio 25

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a = 0  # el minimo
b = 4  # el maximo

# a) Grafique la función de densidad de probabilidad (PDF)
x = np.linspace(-1, 5, 1000) 
pdf = uniform.pdf(x, loc=a, scale=b-a)

plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='f(x) = 0.25', color='blue')
plt.fill_between(x, pdf, alpha=0.3)
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.xlabel('Distancia (km)')
plt.ylabel('Densidad de probabilidad')
plt.xlim(-1, 5)
plt.ylim(-0.05, 0.3)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='red', linestyle='--', label='x=0')
plt.axvline(4, color='red', linestyle='--', label='x=4')
plt.legend()
plt.grid()
plt.show()

# b) Hallar y graficar la función de distribución acumulada (CDF)
cdf = uniform.cdf(x, loc=a, scale=b-a)

plt.figure(figsize=(12, 6))
plt.plot(x, cdf, label='F(x)', color='orange')
plt.title('Función de Distribución Acumulada (CDF)')
plt.xlabel('Distancia (km)')
plt.ylabel('Probabilidad acumulada')
plt.xlim(-1, 5)
plt.ylim(-0.05, 1.05)
plt.axhline(1, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='red', linestyle='--', label='x=0')
plt.axvline(4, color='red', linestyle='--', label='x=4')
plt.legend()
plt.grid()
plt.show()

# c) Hallar la probabilidad de que se produzca una emergencia como máximo a 1 km
prob_max_1km = uniform.cdf(1, loc=a, scale=b-a)
print(f'Probabilidad de que se produzca una emergencia como máximo a 1 km: {prob_max_1km:.4f}')

# d) Hallar la probabilidad de que se produzca una emergencia a más de 1.5 km de la base
prob_less_than_0_5km = uniform.cdf(0.5, loc=a, scale=b-a)  # P(X < 0.5)
prob_more_than_3_5km = 1 - uniform.cdf(3.5, loc=a, scale=b-a)  # P(X > 3.5)

# Sumar ambas probabilidades
prob_more_than_1_5km_from_base = prob_less_than_0_5km + prob_more_than_3_5km
print(f'Probabilidad de que se produzca una emergencia a más de 1.5 km de la base: {prob_more_than_1_5km_from_base:.4f}')
```
