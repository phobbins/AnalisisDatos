### Ejercicio 24

```python
# Definir los parámetros de la distribución uniforme
a = 0  # Límite inferior del intervalo
b = 2  # Límite superior del intervalo

# Función de distribución acumulada (FDC)
def FDC(x):
    if x < a:
        return 0
    elif a <= x <= b:
        return 0.5 * x  # F(x) = 0.5x en el intervalo [0, 2]
    else:
        return 1

# Calcular la probabilidad de que la grieta ocurra entre 0.5 km y 1.5 km
x1 = 0.5
x2 = 1.5

# Probabilidad P(0.5 <= x <= 1.5)
probabilidad = FDC(x2) - FDC(x1)

# Mostrar resultados
print(f"Función de distribución acumulada F(x) para x = 0.5: {FDC(0.5)}")
print(f"Función de distribución acumulada F(x) para x = 1.5: {FDC(1.5)}")
print(f"Probabilidad de que la grieta ocurra entre 0.5 km y 1.5 km: {probabilidad}")
```

### Ejercicio 25

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a = 0  # el minimo
b = 4  # el maximo

# a) Grafique la función de densidad de probabilidad (PDF)
x = np.linspace(-1, 5, 1000) 
pdf = uniform.pdf(x, loc=a, scale=b-a)

plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='f(x) = 0.25', color='blue')
plt.fill_between(x, pdf, alpha=0.3)
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.xlabel('Distancia (km)')
plt.ylabel('Densidad de probabilidad')
plt.xlim(-1, 5)
plt.ylim(-0.05, 0.3)
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='red', linestyle='--', label='x=0')
plt.axvline(4, color='red', linestyle='--', label='x=4')
plt.legend()
plt.grid()
plt.show()

# b) Hallar y graficar la función de distribución acumulada (CDF)
cdf = uniform.cdf(x, loc=a, scale=b-a)

plt.figure(figsize=(12, 6))
plt.plot(x, cdf, label='F(x)', color='orange')
plt.title('Función de Distribución Acumulada (CDF)')
plt.xlabel('Distancia (km)')
plt.ylabel('Probabilidad acumulada')
plt.xlim(-1, 5)
plt.ylim(-0.05, 1.05)
plt.axhline(1, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='red', linestyle='--', label='x=0')
plt.axvline(4, color='red', linestyle='--', label='x=4')
plt.legend()
plt.grid()
plt.show()

# c) Hallar la probabilidad de que se produzca una emergencia como máximo a 1 km
prob_max_1km = uniform.cdf(1, loc=a, scale=b-a)
print(f'Probabilidad de que se produzca una emergencia como máximo a 1 km: {prob_max_1km:.4f}')

# d) Hallar la probabilidad de que se produzca una emergencia a más de 1.5 km de la base
prob_less_than_0_5km = uniform.cdf(0.5, loc=a, scale=b-a)  # P(X < 0.5)
prob_more_than_3_5km = 1 - uniform.cdf(3.5, loc=a, scale=b-a)  # P(X > 3.5)

# Sumar ambas probabilidades
prob_more_than_1_5km_from_base = prob_less_than_0_5km + prob_more_than_3_5km
print(f'Probabilidad de que se produzca una emergencia a más de 1.5 km de la base: {prob_more_than_1_5km_from_base:.4f}')
```
