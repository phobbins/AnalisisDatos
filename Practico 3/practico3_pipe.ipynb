```python
# Reingresando el código debido al reinicio del entorno

# Datos de la muestra
estados = {
    "Alabama": {"poblacion": 4500752, "ingreso_per_capita": 26338},
    "Georgia": {"poblacion": 8684715, "ingreso_per_capita": 29442},
    "Illinois": {"poblacion": 12653544, "ingreso_per_capita": 33690},
    "Indiana": {"poblacion": 6195643, "ingreso_per_capita": 28783},
    "Nueva York": {"poblacion": 19190115, "ingreso_per_capita": 36574},
    "Pensilvania": {"poblacion": 12365455, "ingreso_per_capita": 31998},
    "Tennessee": {"poblacion": 5841748, "ingreso_per_capita": 28455}
}

# Calcular la renta personal media per cápita ponderada
total_poblacion = sum(estado["poblacion"] for estado in estados.values())
renta_per_capita_ponderada = sum(estado["poblacion"] * estado["ingreso_per_capita"] for estado in estados.values()) / total_poblacion

print(f"La renta media per capita ponderada es de {renta_per_capita_ponderada}")
```
