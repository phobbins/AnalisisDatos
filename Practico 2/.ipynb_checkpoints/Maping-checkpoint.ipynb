```python
import pandas as pd
import os

df = pd.read_csv('netflix.csv')
df = df.dropna(subset=['country'])


print(df.columns)

directorio = 'graficosMapa'
if not os.path.exists(directorio):
    os.makedirs(directorio)
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from cartopy import crs as ccrs
from collections import Counter

# Supongamos que tienes un DataFrame df con los datos de pelculas
# Asegrate de que 'country' contenga los nombres de los pases de origen
def procesar_paises(pais_str):
    if isinstance(pais_str, str):
        return [p.strip() for p in pais_str.split(',')]
    return []

# Desglosar las coproducciones
df['country'] = df['country'].apply(procesar_paises)

# Contar la cantidad de pelculas por pas
paises_contados = Counter([country for lista_paises in df['country'] for country in lista_paises])

# Convertir los conteos en un DataFrame
df_paises_contados = pd.DataFrame(paises_contados.items(), columns=['country', 'conteo'])

print(df_paises_contados)

# Ruta al archivo .shp (suponiendo que est en el mismo directorio que el notebook)
shapefile_path = 'ne_110m_admin_0_countries.shp'

# Cargar datos geogrficos de los pases
world = gpd.read_file(shapefile_path)

# Verificar las columnas del GeoDataFrame
print("Columnas disponibles en el GeoDataFrame:")
print(world.columns)  # Verifica qu columna contiene los nombres de los pases

# Asegrate de que la columna que contiene los nombres de los pases en el GeoDataFrame coincida
# con la columna en el DataFrame de conteos
# Actualiza 'ADMIN' con el nombre correcto de la columna

# Por ejemplo, si la columna que contiene los nombres de los pases es 'ADMIN', cambia 'name' a 'ADMIN'
nombre_columna_paises = 'SOVEREIGNT'  # Reemplaza 'ADMIN' con el nombre correcto basado en la impresin anterior

# Unir el DataFrame de pelculas con los datos geogrficos
world = world.merge(df_paises_contados, how='left', left_on=nombre_columna_paises, right_on='country')

# Rellenar los pases sin pelculas con un valor de 0
world['conteo'] = world['conteo'].fillna(0)

# Inferir los tipos de datos para evitar la advertencia futura
world['conteo'] = world['conteo'].infer_objects()

# Crear la visualizacin en un mapa usando Cartopy
fig, ax = plt.subplots(1, 1, figsize=(15, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# Plotea el mapa coloreando los pases de acuerdo al conteo de pelculas
world.plot(column='conteo', ax=ax, legend=True,
           legend_kwds={'label': "Cantidad de Pelculas por Pas",
                        'orientation': "horizontal"},
           cmap='OrRd')

# Aadir detalles al mapa
ax.set_title('Distribucin de Pelculas por Pas', fontsize=15)
ax.coastlines()
plt.show()

```
